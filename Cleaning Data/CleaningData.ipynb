```python
import pandas as pd
```

```python
df = pd.read_csv("Predict_Hair_Fall.csv")
```

```python
df.head()
```

```python
# Dropping the ID column as there is an existing index

df = df.drop('Id', axis=1)
```

```python
# Renaming the columns that had a single space at the end to make future coding easier

df = df.rename(columns={'Weight Loss ': 'Weight Loss', 'Poor Hair Care Habits ': 'Poor Hair Care Habits'})
```

```python
# converting Yes and No to 1 and 0 to make a binary columns for the model

df[['Genetics', 'Hormonal Changes', 'Poor Hair Care Habits', 'Environmental Factors', 'Smoking', 'Weight Loss']] = \
    df[['Genetics', 'Hormonal Changes', 'Poor Hair Care Habits', 'Environmental Factors', 'Smoking', 'Weight Loss']].replace({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
df.to_csv("CleanedHair.csv", index = False)
```
